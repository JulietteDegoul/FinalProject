```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# download stopwords and wordnet lemmatizer if necessary
nltk.download('stopwords')
nltk.download('wordnet')

# create a list to store the preprocessed documents
p_corpus = []

# initialize the stop words and lemmatizer objects
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# loop over each document in the corpus
for text in df['text']:
    
    # tokenize the document into sentences
    sentences = nltk.sent_tokenize(text)
    
    # loop over each sentence in the document
    for sentence in sentences:
        
        # tokenize the sentence into words
        words = nltk.word_tokenize(sentence)
        
        # remove stop words and lemmatize the remaining words
        filtered_words = [lemmatizer.lemmatize(w.lower()) for w in words if not w.lower() in stop_words]
        
        # add the preprocessed sentence to the processed corpus
        p_corpus.append(' '.join(filtered_words))
```

```python
for i in range(len(p_corpus)):
    p_corpus[i] = p_corpus[i].lower()
    p_corpus[i] = re.sub(r'\W+',' ',p_corpus[i]) # Replace everything non-alphanumeric by ' '
    p_corpus[i] = re.sub(r'\d+',' ',p_corpus[i]) # Replace one or more digits by  ' '
    p_corpus[i] = re.sub(r'([a-z0-9+._-]+@[a-z0-9+._-]+\.[a-z0-9+_-]+)'," ", p_corpus[i]) # Replace e-mails by ' '
    # Replace urls by ''
    p_corpus[i] = re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', ' ' , p_corpus[i]) 
    # Replace html tags by ''
    p_corpus[i] = BeautifulSoup(p_corpus[i], 'html.parser').get_text().strip()
    p_corpus[i] = re.sub(r'\s+',' ',p_corpus[i]) # Replace one or more whitespaces by  ' '
```

```python
from keras.callbacks import EarlyStopping
import pandas as pd
import numpy as np
from keras.preprocessing.text import Tokenizer
#from keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Dense, Dropout, Embedding, LSTM, GlobalMaxPooling1D

# Load the data
#df = pd.read_csv('your_data.csv')

# Separate the text and labels
#corpus = df['text'].values
train_label = df['is_success'].values

# Create a tokenizer object
tokenizer = Tokenizer()

# Fit the tokenizer on the corpus
tokenizer.fit_on_texts(p_corpus)

# Convert text to sequence of tokens
sequences = tokenizer.texts_to_sequences(p_corpus)

# Pad sequences to have equal length
max_len = max([len(seq) for seq in sequences])
padded_sequences = pad_sequences(sequences, maxlen=max_len)

# Make sure the number of samples is the same for both input and target data
num_samples = min(len(padded_sequences), len(train_label))
padded_sequences = padded_sequences[:num_samples]
train_label = train_label[:num_samples]

# Define the model
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=128, input_length=max_len))
model.add(LSTM(units=64, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(units=1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Define early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=2)

# Train the model
model.fit(padded_sequences, train_label, epochs=10, verbose=1, validation_split=0.2, callbacks=[early_stopping], batch_size = 32)

```
